Example: name, age, balance, etc.

These are the actions/behaviors that an object can perform.They usually use or

Example: deposit_money(), withdraw_money(), display_info(), etc


###2. Object:

```python
# Function vs Method Example

#Function ----> outside, general for all
# Method -----> inside of a class, specific for class only

# Function example
my_list = [1, 2, 3, 4]

print(len(my_list))   # len() is a function, not tied to any object

my_list.append(5)   # append() is a method of the class 'list' which have an object name as my_list
print(my_list)
```

```python
# Project ATM

class ATM:

  def __init__(self):

    #instance variable ---> A variable for which value is different for different object
    self.pin=""
    self.balance=0
    self.menu()


  def menu(self):
    user_input=input("""
    Hello! How would you like to proceed?

    1. Enter 1 to create pin
    2. Enter 2 to deposit
    3. Enter 3 to withdraw
    4. Enter 4 to check balance
    5. Enter 5 to exit

        """)

    if user_input == "1":
      self.create_pin()

    elif user_input=="2":
      self.deposit()

    elif user_input=="3":
      self.withdraw()

    elif user_input=="4":
        self.check_balance()

    else:
      print("Bye!")

  def create_pin(self):
    self.pin=input("Enter your pin ")
    print("Pin set successfully")


  def deposit(self):
    temp=input("Enter your pin ")
    if temp==self.pin:
      amount=int(input("Enter your amount "))
      self.balance=self.balance+amount
      print("Deposit Successful")
    else:
      print("Invalid Pin")

  def withdraw(self):
    temp=input("Enter your pin ")
    if temp==self.pin:
      amount=int(input("Enter your amount "))
      if self.balance>amount:
        self.balance=self.balance-amount
        print("Withdraw Successful")
      else:
        print("Insufficient Balance")
    else:
      print("Invalid Pin")

  def check_balance(self):
    temp=input("Enter your pin ")
    if temp==self.pin:
      print(self.balance)
    else:
      print("Invalid Pin")
```

# Encapsulation

Encapsulation means wrapping variables and methods together in a class and hiding the internal details, so users interact with the object only through defined methods, not by directly modifying its data.


### Use Case:

it is used where we have to hide the some internals of data for Example:

* A patient’s medical history should not be directly accessible. Instead, only authorized methods (doctor’s view, patient’s view) can show parts of it.

* Car speed (__speed) shouldn’t be set directly. It should be controlled via methods like accelerate() or brake().

* You can’t directly add yourself to someone’s friend list on facebook.Instead, there’s a controlled process: send_request() → accept_request().


```python
# Encapsulatin ----> make privte the instances

class Employee:
    def __init__(self, name, salary):
        self.name = name          # public attribute
        self.__salary = salary    # private attribute
```

```python
emp = Employee("Fedrick", 50000)
```

You notice that now the instance which make us private are not showing, this is encapsulation

```python
# Public ----> shows to everyone
emp.name
```

```python
# Private ----> Shows error because of access restrictions

emp.__salary
```

# Inheritance

Inheritance is an Object-Oriented Programming (OOP) principle where a class (called the child/derived class) can acquire the properties (attributes) and behaviors (methods) of another class (called the parent/base class).

* Parent Class : A class from where child class access attributes and methods.
* Child Class : A class which access the attributes and methods of parent calss

### UseCase:

* Common features (login, username) are in User.Specific roles extend those features without duplicating code.

```python
# Inheritance Exmple

# Parent class
class User:

  def login(self):
    print("Login")

  def register(self):
    print("Register")

# Child class  --->  child class access al the things of parent class
class Student(User):

  def enroll(self):
    print("Enroll")

  def review(self):
    print("Review")
```

```python
s1=Student()
```

```python
# S1 is the object ofchild class that can access their own attributes & methods as well as the parent class attributes & methods

# child class methods
s1.enroll()
s1.review()

# parent class methods
s1.login()
s1.register()
```

The super() method in is used to call a method (usually __init__ or any other function) from the parent class inside the child class.

```python
# Example of inheritance with super method

# Parent class
class car():
    def __init__(self,seats,tyres,stairing,brake,horn,car_name):
        self.seats=seats
        self.tyres=tyres
        self.stairing=stairing
        self.brake=brake
        self.horn=horn
        self.car_name=car_name

    def speed(self):
        print(f"The {self.car_name} is running ")

    def describe_car(self):
        print(f"""
        Car Model: {self.car_name}
        Total Seats: {self.seats}
        Tyres: {self.tyres}
        Horn Sound: {self.horn}
        Stairing Side: {self.stairing}
        Break: {self.brake}""")


# Child class
class electric_car(car):
  def __init__(self,make,model,year,engine,seats,tyres,stairing,brake,horn,car_name):

      # super() method
      super().__init__(seats,tyres,stairing,brake,horn,car_name)
      self.make=make
      self.model=model
      self.year=year
      self.engine=engine


  def describe_electric_car(self):
      print(f"""
      Company: {self.make}
      Model Name: {self.model}
      Year: {self.year}
      Engine Specification: {self.engine}""")
```

```python
e1=electric_car('Tesla','ES01',2021,'Erc20',4,4,'Right','jix','light','jinusa')
```

```python
# parent methods
e1.speed()
e1.describe_car()

# child method
e1.describe_electric_car()
```

# Polymorphism

Polymorphism means "many forms".

In Object-Oriented Programming (OOP), polymorphism allows the same function or method name to behave differently depending on the object or data type that calls it.

In simple words: One name, multiple behaviors.

### Use Case:

Payment Methods

Action: make payment.

With a credit card → swipe the card.

With cash → give money.

With mobile wallet → scan QR code.

One interface (payment), different implementations.

```python
# Example of polymorphism
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")
```

```python
# Usage ----> man forms f same method (sound)
animals = [Dog(), Cat(), Animal()]
for a in animals:
    a.sound()
```

```python
# Plymorphism Example

class car():
    def __init__(self,seats,tyres,stairing,brake,horn,car_name):
        self.seats=seats
        self.tyres=tyres
        self.stairing=stairing
        self.brake=brake
        self.horn=horn
        self.car_name=car_name

    def speed(self):
        print(f"The {self.car_name} is running ")

    def sound(self):
        print(f"{self.car_name} sound is very loud")

    def describe_car(self):
        print(f"""
        Car Model: {self.car_name}
        Total Seats: {self.seats}
        Tyres: {self.tyres}
        Horn Sound: {self.horn}
        Stairing Side: {self.stairing}
        Break: {self.brake}""")

```

```python
c1=car(4,4,'left','gic','loud','aqua')

c1.sound()
```

```python
class electric_car(car):
    def __init__(self,make,model,year,engine,seats,tyres,stairing,brake,horn,car_name):

        super().__init__(seats,tyres,stairing,brake,horn,car_name)
        self.make=make
        self.model=model
        self.year=year
        self.engine=engine

    def sound(self):
        print(f"{self.car_name} is running silently")


    def describe_electric_car(self):
        print(f"""
        Company: {self.make}
        Model Name: {self.model}
        Year: {self.year}
        Engine Specification: {self.engine}""")
```

```python
e1=electric_car('Honda','wq12',2018,'P201A',4,4,'left','huz','light','city')
```

```python
e1.sound() # same method but different working
```

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def move(self):
        print("Drive!")

class Boat:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def move(self):
        print("Sail!")

class Plane:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def move(self):
        print("Fly!")

# Create objects of different types
car1 = Car("Ford", "Mustang")
boat1 = Boat("Ibiza", "Touring 20")
plane1 = Plane("Boeing", "747")

# Put the objects in a list
vehicles = [car1, boat1, plane1]

# Iterate through the list and call the move() method
for x in vehicles:
    print(x.brand, x.model)
    x.move()
```

# Abstraction

Abstraction means hiding the implementation details and showing only the essential features to the user.

The user doesn’t need to know how things work internally, they only use the interface.

In Python, abstraction is mostly achieved using abstract classes and abstract methods (e.g random module).


### Use Case:

* You use the steering wheel, brake, and accelerator.You don’t know how the engine works internally, how the hydraulic brake system functions, or how the gearbox rotates. The driver only sees the necessary parts, while the internal implementation is hidden.

```python
# Examle of abstraction
import math # abstracted library

print(math.sqrt(16))
```

```python
# Example of abstraction
import random  # abtracted library

num = random.randint(1, 10)
print(num)
```
